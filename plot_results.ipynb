```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_results(fname):
    plt.figure(figsize=(12,5)) if not plt.get_fignums() else plt.clf()
    res = pd.read_csv(fname)
    for r in res:
        if "loss" in r:
            plt.subplot(1,2,1)
            plt.plot(res[r], label=r)
        elif "metric" in r:
            plt.subplot(1,2,2)
            plt.plot(res[r], label=r)
    plt.subplot(1,2,1)
    plt.yscale("log")
    plt.legend()
    plt.subplot(1,2,2)
    plt.legend()
    plt.show()
```

```python
plot_results("runs/detect/train/results.csv")
```

```python
plot_results("runs/zaxis/train13/results.csv")
```

```python
plot_results("runs/zaxis/train15/results.csv")
```
